# Código para geração de caso do Zumbi Health a partir de um CSV

```python
import pandas as pd
import numpy as np
```

```python
def CSVtoCase(filename):
    
    patient = ""
    
    df = pd.read_csv(filename)
    
    cases = df.values
    
    for c, value in enumerate(cases):
        print("=== Caso {} (case) ===".format(c+1))
        patient = value[0]
        print("![{}]({}.png)".format(patient, c+1))
        print("{} se apresentou na clínica para ser atendido. O que você deseja fazer?".format(patient))
        symptoms = []
        for v in value[1:-1]:
            if v == 0 :
                symptoms += ["ok"]
            else:
                symptoms += ["problema"]
        print("++ Ver movimentos -> Paralisia {}".format(symptoms[0]))
        print("++ Examinar língua -> Lingua {}".format(symptoms[1]))
        print("++ Analisar as mãos -> Dedos {}".format(symptoms[2]))
        print("++ Examinar membros -> Membros {}".format(symptoms[3]))
        print("++ Sente dor? -> Dor {}".format(symptoms[4]))
        print("++ Ver humor -> Raiva {}".format(symptoms[5]))
        print("=== Tratamento 1 ===")
        print("![Treatment](images/doctor.png)")
        print("Qual é o tratamento que você deseja aplicar?")
        diagnosis = []
        d = value[-1]
        if d == 1:
            diagnosis += ["certo"]
        else:
            diagnosis += ["errado"]
        if d == 2:
            diagnosis += ["certo"]
        else:
            diagnosis += ["errado"]
        if d == 3:
            diagnosis += ["certo"]
        else:
            diagnosis += ["errado"]
        if d == 4:
            diagnosis += ["certo"]
        else:
            diagnosis += ["errado"]
        labelDiagnosis = ["", "Zulumbriga", "Bacterias", "Virus", "Brigas"]
        
        print("++ Tratar Zulumbriga -> {0} {1} \n++ Tratar Bactérias -> {0} {2} \n++ Tratar Vírus -> {0} {3} \n++ Tratar Brigas -> {0} {4}".format(labelDiagnosis[d],diagnosis[0],diagnosis[1],diagnosis[2],diagnosis[3]))
        
        print("\n")
        
    return cases
        
        
# === Caso 1 (case) ===
# ![Rot_Donnadd](images/1.png)
# Rot Donnadd se apresentou na clínica para ser atendido. O que você deseja fazer?
# ++ Examinar língua -> Lingua amarela
# ++ Ver movimentos -> Paralisia
# ++ Analisar as mãos -> Dedos ok
# ++ Examinar membros -> Membros ok
# ++ Sente dor? -> Sem dor
# ++ Ver humor -> Sem raiva
# ++ Aplicar tratamento -> Tratamento 1
# === Tratamento 1 ===
# ![Treatment](images/doctor.png)
# Qual é o tratamento que você deseja aplicar?
# ++ Trata Bactérias -> Bacterias ok
# ++ Trata Vírus -> Bacterias errado
# ++ Trata Zulumbriga -> Bacterias errado
# ++ Trata Brigas -> Bacterias errado
```
