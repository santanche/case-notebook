# Combing customized element, shadow DOM and template
Simple component using: customized element, shadow DOM and template. [See in JSFiddle](https://jsfiddle.net/santanche/9jcag0ob/)

```python
%%HTML
<!DOCTYPE html>
<html>
  <style>
    body {
      font-family: "Trebuchet MS", Helvetica, sans-serif;
      color: black;
    }
  </style>
  <template id="template-state-selector">
    <style>
      .dcc_state_selector {
        background-color: lightgrey;
        text-decoration: none;
      }

      .dcc_state_selector:hover {
        color: white;
      }
    </style>
    <span id="inner-dcc" class="dcc_state_selector"></span>
  </template>
  <script>
    class DCCStateSelector extends HTMLElement {
      constructor() {
        super();
        this._content = this.innerHTML;

        var template = document.getElementById("template-state-selector").content;
        this._shadow = this.attachShadow({mode: 'open'});
        this._shadow.appendChild(template.cloneNode(true));
      }

      static get observedAttributes() {
        return ["content"];
      }

      createdCallback() {
        this._updateRendering();
      }

      attributeChangedCallback(name, oldValue, newValue) {
        this._updateRendering();
      }

      connectedCallback() {
        this._updateRendering();
      }

      get content() {
        return this.getAttribute("content");
      }

      set content(newContent) {
        this.setAttribute("content", newContent);
      }

      _updateRendering() {
        var innerDCC = this._shadow.querySelector("#inner-dcc");
        if (innerDCC != null)
          innerDCC.innerHTML = this._content;
      }
    }

    customElements.define("dcc-state-selector", DCCStateSelector);
</script>
<head>
  <title>DCC with template</title>
</head>
<body>
      This is the <dcc-state-selector>selector 1</dcc-state-selector>
      and this is the <dcc-state-selector>selector 2</dcc-state-selector>.
</body>
</html>
```
